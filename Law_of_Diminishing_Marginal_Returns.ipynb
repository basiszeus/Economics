```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Data
workers = [1, 2, 3, 4, 5, 6]
total_output = [10, 22, 32, 38, 41, 42]

# Calculate marginal output
marginal_output = [total_output[0]] + [
    total_output[i] - total_output[i-1] for i in range(1, len(total_output))
]

# Plotting
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(workers, total_output, marker='o')
plt.title("Total Output vs Workers")
plt.xlabel("Number of Workers")
plt.ylabel("Total Output")

plt.subplot(1, 2, 2)
plt.plot(workers, marginal_output, marker='o', color='orange')
plt.title("Marginal Output vs Workers")
plt.xlabel("Number of Workers")
plt.ylabel("Marginal Output")

plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame({
    'Workers': workers,
    'Total Output': total_output,
    'Marginal Output': marginal_output
})

df
```
