Supply Curve
</h1><center/>

```python
def plot_supply_curve(prices, quantities_supplied):
    plt.figure(figsize=(8, 6))
    plt.plot(quantities_supplied, prices, marker='o', linestyle='-', color='b', label='Supply Curve')
    plt.ylabel('Price ($)')
    plt.xlabel('Quantity Supplied')
    plt.title('Supply Curve')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
prices = [2, 3, 4, 5, 6]
quantities_supplied = [200, 400, 600, 800, 1000]

plot_supply_curve(prices, quantities_supplied)
```

Percentage Changes
</h1><center/>

```python
def calculate_percentage_change(old_value, new_value):
    if old_value == 0:
        return "Error: Old value cannot be zero."
    else:
        percentage_change = ((new_value - old_value) / old_value) * 100
        if percentage_change > 0:
            print(f"Increased by {percentage_change}%")
        elif percentage_change < 0:
            print(f"Decreased by {-percentage_change}%")
        else:
            print("No change")
```

```python
calculate_percentage_change(50,70)
```

Demand Curve Shift
</h1><center/>

Let's have an example and solve it using Python. We suppose there is a sudden increase in consumer preference for Vitamin D due to the season, leading to higher demand at the same price.<br><br>
The price of a pack of Vitamin D supplements. Initially, the price was regulated by the government at $50, and the quantity demanded by pharmacies in a region was 50 packs. However, due to seasonal changes and increased awareness about health benefits, the demand increased to 70 packs while the price remained constant.

```python
def plot_supply_shift(prices, quantity_supplied, shift_amount, increase=True):
    """
    Plot the original and shifted supply curves.
    
    Parameters:
    - prices (numpy array): Array of price levels.
    - quantity_supplied (numpy array): Array of quantities supplied corresponding to each price level.
    - shift_amount (float): The amount by which the supply curve shifts.
    - increase (bool): If True, shift the supply curve to the right (increase supply). If False, shift the supply curve to the left (decrease supply).
    """
    if increase:
        quantity_supplied_new = quantity_supplied + shift_amount
    else:
        quantity_supplied_new = quantity_supplied - shift_amount

    # Keeping price constant
    price_constant = prices[len(prices) // 2]  # Choose the middle price for constant price visualization
    quantity_supplied_at_price = quantity_supplied[len(prices) // 2]
    quantity_supplied_new_at_price = quantity_supplied_new[len(prices) // 2]

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.plot(quantity_supplied, prices, label='Original Supply', color='blue')
    plt.plot(quantity_supplied_new, prices, label='Shifted Supply', color='green', linestyle='--')
    plt.axhline(y=price_constant, color='gray', linestyle=':', label=f'Constant Price = {price_constant}')
    plt.scatter([quantity_supplied_at_price, quantity_supplied_new_at_price], [price_constant, price_constant], color='red')
    plt.annotate(f'Q1 = {quantity_supplied_at_price:.2f}', (quantity_supplied_at_price, price_constant), textcoords="offset points", xytext=(-15,5), ha='center', color='red')
    plt.annotate(f'Q2 = {quantity_supplied_new_at_price:.2f}', (quantity_supplied_new_at_price, price_constant), textcoords="offset points", xytext=(15,5), ha='center', color='red')
    plt.xlabel('Quantity Supplied')
    plt.ylabel('Price')
    plt.title('Shift in Supply Curve at Constant Price')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
prices = np.linspace(0, 100, 100) #100 prices ranging from 0 to 100
quantity_supplied = 2 * prices
```

```python
# Increase in supply
plot_supply_shift(prices, quantity_supplied, shift_amount=50, increase=True)
```

```python
# Decrease in supply
plot_supply_shift(prices, quantity_supplied, shift_amount=50, increase=False)
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Price Elasticity of Supply (PES)
</h1><center/>

```python
import numpy as np
import matplotlib.pyplot as plt

def analyze_price_elasticity(price_initial, price_final, quantity_initial, quantity_final):
    percent_change_price = (price_final - price_initial) / price_initial * 100
    percent_change_quantity = (quantity_final - quantity_initial) / quantity_initial * 100

    elasticity = percent_change_quantity / percent_change_price

    analysis = (
        f"Initial Price: ${price_initial}\n"
        f"Final Price: ${price_final}\n"
        f"Initial Quantity Supplied: {quantity_initial}\n"
        f"Final Quantity Supplied: {quantity_final}\n"
        f"\n"
        f"Percentage change in price: {percent_change_price:.2f}%\n"
        f"Percentage change in quantity supplied: {percent_change_quantity:.2f}%\n"
        f"Price elasticity of supply (PES): {elasticity:.2f}\n"
    )
    
    if elasticity > 1:
        analysis += "The supply is elastic, indicating that suppliers are sensitive to price changes."
    elif elasticity < 1:
        analysis += "The supply is inelastic, indicating that suppliers are less sensitive to price changes."
    else:
        analysis += "The supply has unitary elasticity, indicating that the percentage change in quantity supplied is equal to the percentage change in price."

    # Create the supply curve
    price = np.linspace(price_initial, price_final, 100)
    quantity_supplied = quantity_initial + (quantity_final - quantity_initial) / (price_final - price_initial) * (price - price_initial)

    plt.figure(figsize=(10, 6))
    plt.plot(quantity_supplied, price, label='Supply Curve', color='blue')
    plt.scatter([quantity_initial, quantity_final], [price_initial, price_final], color='red')
    plt.annotate(f'Initial Point\nQ1 = {quantity_initial}\nP1 = ${price_initial}', 
                 (quantity_initial, price_initial), textcoords="offset points", xytext=(-40,-10), ha='center', color='red')
    plt.annotate(f'Final Point\nQ2 = {quantity_final}\nP2 = ${price_final}', 
                 (quantity_final, price_final), textcoords="offset points", xytext=(40,10), ha='center', color='red')
    plt.xlabel('Quantity Supplied')
    plt.ylabel('Price')
    plt.title('Price Elasticity of Supply')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(analysis)
```

Let's say we are analyzing the supply of widgets in a market. Initially, the price of a widget is 50, and the quantity supplied at this price is 500 units. Due to some market changes, the price of widgets increases to 60, and as a result, the quantity supplied increases to 700 units. We want to analyze the price elasticity of supply to understand how sensitive the suppliers are to the change

```python
price_initial = 50
price_final = 60
quantity_initial = 500
quantity_final = 700

analyze_price_elasticity(price_initial, price_final, quantity_initial, quantity_final)
```
