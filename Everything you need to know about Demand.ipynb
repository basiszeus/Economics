Percentage Changes
</h1><center/>

```python
def calculate_percentage_change(old_value, new_value):
    if old_value == 0:
        return "Error: Old value cannot be zero."
    else:
        percentage_change = ((new_value - old_value) / old_value) * 100
        if percentage_change > 0:
            print(f"Increased by {percentage_change}%")
        elif percentage_change < 0:
            print(f"Decreased by {-percentage_change}%")
        else:
            print("No change")
```

```python
calculate_percentage_change(50,70)
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Demand Curve Shift
</h1><center/>

Let's have an example and solve it using Python. We suppose there is a sudden increase in consumer preference for Vitamin D due to the season, leading to higher demand at the same price.<br><br>
The price of a pack of Vitamin D supplements. Initially, the price was regulated by the government at $50, and the quantity demanded by pharmacies in a region was 50 packs. However, due to seasonal changes and increased awareness about health benefits, the demand increased to 70 packs while the price remained constant.

```python
price = np.linspace(0, 100, 100)
quantity_demanded = 100 - price

# New demand curve (shifted to the right)
quantity_demanded_new = 120 - price

# Keeping price constant
price_constant = 50
quantity_demanded_at_price = 100 - price_constant
quantity_demanded_new_at_price = 120 - price_constant

# Vizualisation

plt.figure(figsize=(10, 6))
plt.plot(quantity_demanded, price, label='Original Demand', color='blue')
plt.plot(quantity_demanded_new, price, label='Increased Demand', color='green', linestyle='--')
plt.axhline(y=price_constant, color='gray', linestyle=':', label=f'Constant Price = {price_constant}')
plt.scatter([quantity_demanded_at_price, quantity_demanded_new_at_price], [price_constant, price_constant], color='red')
plt.annotate(f'Q1 = {quantity_demanded_at_price}', (quantity_demanded_at_price, price_constant), textcoords="offset points", xytext=(-15,5), ha='center', color='red')
plt.annotate(f'Q2 = {quantity_demanded_new_at_price}', (quantity_demanded_new_at_price, price_constant), textcoords="offset points", xytext=(15,5), ha='center', color='red')
plt.xlabel('Quantity Demanded')
plt.ylabel('Price')
plt.title('Increase in Demand at Constant Price')
plt.legend()
plt.grid(True)
plt.show()
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Price Elasticity of Demand (PED)
</h1><center/>

```python
def analyze_price_elasticity(price_initial, price_final, quantity_initial, quantity_final):
    # Calculate percentage changes
    percent_change_price = (price_final - price_initial) / price_initial * 100
    percent_change_quantity = (quantity_final - quantity_initial) / quantity_initial * 100

    # Calculate price elasticity of demand
    elasticity = percent_change_quantity / percent_change_price

    # Analysis
    analysis = (
        f"Initial Price: ${price_initial}\n"
        f"Final Price: ${price_final}\n"
        f"Initial Quantity Demanded: {quantity_initial}\n"
        f"Final Quantity Demanded: {quantity_final}\n"
        f"\n"
        f"Percentage change in price: {percent_change_price:.2f}%\n"
        f"Percentage change in quantity demanded: {percent_change_quantity:.2f}%\n"
        f"Price elasticity of demand (PED): {elasticity:.2f}\n"
    )
    
    if elasticity > 1:
        analysis += "The demand is elastic, indicating that consumers are sensitive to price changes."
    elif elasticity < 1:
        analysis += "The demand is inelastic, indicating that consumers are less sensitive to price changes."
    else:
        analysis += "The demand has unitary elasticity, indicating that the percentage change in quantity demanded is equal to the percentage change in price."

    # Data for demand curve
    price = np.linspace(price_initial - 1, price_final + 1, 100)
    quantity_demanded = quantity_initial - (quantity_initial - quantity_final) / (price_final - price_initial) * (price - price_initial)

    # Plotting the demand curve and points
    plt.figure(figsize=(10, 6))
    plt.plot(quantity_demanded, price, label='Demand Curve', color='blue')
    plt.scatter([quantity_initial, quantity_final], [price_initial, price_final], color='red')
    plt.annotate(f'Initial Point\nQ1 = {quantity_initial}\nP1 = ${price_initial}', 
                 (quantity_initial, price_initial), textcoords="offset points", xytext=(-40,-10), ha='center', color='red')
    plt.annotate(f'Final Point\nQ2 = {quantity_final}\nP2 = ${price_final}', 
                 (quantity_final, price_final), textcoords="offset points", xytext=(40,10), ha='center', color='red')
    plt.xlabel('Quantity Demanded')
    plt.ylabel('Price')
    plt.title('Price Elasticity of Demand')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print analysis
    print(analysis)
```

Let's consider the case of coffee in a small town. The price of a cup of coffee increases from $2 to $2.50. Initially, the quantity demanded was 100 cups per day. After the price increase, the quantity demanded drops to 80 cups per day.

```python
# Example usage
price_initial = 2.00
price_final = 2.50
quantity_initial = 100
quantity_final = 80

analyze_price_elasticity(price_initial, price_final, quantity_initial, quantity_final)
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Cross-Price Elasticity of Demand
</h1><center/>

```python
def cross_price_elasticity_pharma(price_initial_B, price_final_B, quantity_initial_A, quantity_final_A):
    # Calculate percentage changes
    percent_change_price_B = (price_final_B - price_initial_B) / price_initial_B * 100
    percent_change_quantity_A = (quantity_final_A - quantity_initial_A) / quantity_initial_A * 100

    # Calculate cross-price elasticity
    cpe = percent_change_quantity_A / percent_change_price_B

    # Analysis
    analysis = (
        f"Initial Price of Drug B: ${price_initial_B}\n"
        f"Final Price of Drug B: ${price_final_B}\n"
        f"Initial Quantity Demanded of Drug A: {quantity_initial_A} units\n"
        f"Final Quantity Demanded of Drug A: {quantity_final_A} units\n"
        f"\n"
        f"Percentage change in price of Drug B: {percent_change_price_B:.2f}%\n"
        f"Percentage change in quantity demanded of Drug A: {percent_change_quantity_A:.2f}%\n"
        f"Cross-price elasticity of demand (CPE): {cpe:.2f}\n"
    )
    
    if cpe > 0:
        analysis += "The goods are substitutes."
    elif cpe < 0:
        analysis += "The goods are complements."
    else:
        analysis += "The goods are unrelated."

    # Print analysis
    print(analysis)

    # Data for demand curve
    prices_B = [price_initial_B, price_final_B]
    quantities_A = [quantity_initial_A, quantity_final_A]

    # Plotting the change
    plt.figure(figsize=(10, 6))
    plt.plot(prices_B, quantities_A, marker='o', linestyle='-', color='blue')
    plt.xlabel('Price of Drug B')
    plt.ylabel('Quantity Demanded of Drug A')
    plt.title('Cross-Price Elasticity of Demand in Pharmaceuticals')
    plt.grid(True)
    plt.show()
```

```python
# Example usage
price_initial_B = 30.00
price_final_B = 40.00
quantity_initial_A = 1000
quantity_final_A = 1300

cross_price_elasticity_pharma(price_initial_B, price_final_B, quantity_initial_A, quantity_final_A)
```
