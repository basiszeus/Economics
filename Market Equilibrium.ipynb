```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Calculation and vizualisation of Market Equilibruim
</h1><center/>

```python
def calculate_slope_intercept(data):
    prices, quantities = zip(*data)
    slope, intercept, _, _, _ = linregress(prices, quantities)
    return slope, intercept

def plot_curve(prices, quantities, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(prices, quantities, color='blue')
    slope, intercept = calculate_slope_intercept(list(zip(prices, quantities)))
    plt.plot(prices, slope * np.array(prices) + intercept, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
    plt.xlabel('Price')
    plt.ylabel('Quantity')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
    return slope, intercept

def market_equilibrium(demand_data, supply_data):
    # Plot demand and supply curves
    demand_prices, demand_quantities = zip(*demand_data)
    supply_prices, supply_quantities = zip(*supply_data)
    
    demand_slope, demand_intercept = plot_curve(demand_prices, demand_quantities, 'Demand Curve')
    supply_slope, supply_intercept = plot_curve(supply_prices, supply_quantities, 'Supply Curve')
    
    print(f'Demand Curve: Qd = {demand_intercept:.2f} - {abs(demand_slope):.2f}P')
    print(f'Supply Curve: Qs = {supply_intercept:.2f} + {abs(supply_slope):.2f}P')
    
    # Calculate equilibrium
    Pe = (supply_intercept - demand_intercept) / (demand_slope - supply_slope)
    Qe = demand_intercept + demand_slope * Pe
    
    print(f'Equilibrium Price (Pe): {Pe:.2f}')
    print(f'Equilibrium Quantity (Qe): {Qe:.2f}')
    
    # Visualization of market equilibrium
    prices = np.linspace(0, max(demand_prices + supply_prices), 400)
    demand_curve = demand_intercept + demand_slope * prices
    supply_curve = supply_intercept + supply_slope * prices
    
    plt.figure(figsize=(10, 6))
    plt.plot(prices, demand_curve, label='Demand Curve (Qd)')
    plt.plot(prices, supply_curve, label='Supply Curve (Qs)')
    plt.axvline(Pe, color='grey', linestyle='--', label=f'Equilibrium Price (Pe) = {Pe:.2f}')
    plt.axhline(Qe, color='grey', linestyle='--', label=f'Equilibrium Quantity (Qe) = {Qe:.2f}')
    plt.scatter([Pe], [Qe], color='red')
    plt.text(Pe, Qe, f'({Pe:.2f}, {Qe:.2f})', ha='right')
    
    plt.xlabel('Price')
    plt.ylabel('Quantity')
    plt.title('Market Equilibrium')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
demand_data = [(10, 80), (12, 70), (14, 60), (16, 50), (18, 40)]
supply_data = [(10, 30), (12, 40), (14, 50), (16, 60), (18, 70)]

market_equilibrium(demand_data, supply_data)
```

```python
def calculate_slope_intercept(data):
    prices, quantities = zip(*data)
    slope, intercept, _, _, _ = linregress(prices, quantities)
    return slope, intercept

def plot_curve(prices, quantities, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(prices, quantities, color='blue')
    slope, intercept = calculate_slope_intercept(list(zip(prices, quantities)))
    plt.plot(prices, slope * np.array(prices) + intercept, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
    plt.xlabel('Price')
    plt.ylabel('Quantity')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
    return slope, intercept

def market_equilibrium(demand_data, supply_data):
    # Plot demand and supply curves
    demand_prices, demand_quantities = zip(*demand_data)
    supply_prices, supply_quantities = zip(*supply_data)
    
    demand_slope, demand_intercept = plot_curve(demand_prices, demand_quantities, 'Demand Curve')
    supply_slope, supply_intercept = plot_curve(supply_prices, supply_quantities, 'Supply Curve')
    
    print(f'Demand Curve: Qd = {demand_intercept:.2f} - {abs(demand_slope):.2f}P')
    print(f'Supply Curve: Qs = {supply_intercept:.2f} + {abs(supply_slope):.2f}P')
    
    # Calculate equilibrium
    Pe = (supply_intercept - demand_intercept) / (demand_slope - supply_slope)
    Qe = demand_intercept + demand_slope * Pe
    
    print(f'Equilibrium Price (Pe): {Pe:.2f}')
    print(f'Equilibrium Quantity (Qe): {Qe:.2f}')
    
    # Visualization of market equilibrium
    prices = np.linspace(0, max(demand_prices + supply_prices), 400)
    demand_curve = demand_intercept + demand_slope * prices
    supply_curve = supply_intercept + supply_slope * prices
    
    plt.figure(figsize=(10, 6))
    plt.plot(prices, demand_curve, label='Demand Curve (Qd)')
    plt.plot(prices, supply_curve, label='Supply Curve (Qs)')
    plt.axvline(Pe, color='grey', linestyle='--', label=f'Equilibrium Price (Pe) = {Pe:.2f}')
    plt.axhline(Qe, color='grey', linestyle='--', label=f'Equilibrium Quantity (Qe) = {Qe:.2f}')
    plt.scatter([Pe], [Qe], color='red')
    plt.text(Pe, Qe, f'({Pe:.2f}, {Qe:.2f})', ha='right')
    
    plt.xlabel('Price')
    plt.ylabel('Quantity')
    plt.title('Market Equilibrium')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return Pe, Qe

def plot_equilibrium_shift(initial_demand, initial_supply, shifted_demand=None, shifted_supply=None):
    # Calculate initial equilibrium
    print("Initial Equilibrium:")
    initial_pe, initial_qe = market_equilibrium(initial_demand, initial_supply)
    
    # Check if we have a shifted demand or supply curve
    if shifted_demand is not None:
        print("\nNew Equilibrium with Shifted Demand:")
        # Calculate and plot new equilibrium with shifted demand
        shifted_pe_d, shifted_qe_d = market_equilibrium(shifted_demand, initial_supply)
        
        # Plot equilibrium shift
        plt.figure(figsize=(10, 6))
        plt.plot([initial_pe, shifted_pe_d], [initial_qe, shifted_qe_d], marker='o')
        plt.scatter([initial_pe, shifted_pe_d], [initial_qe, shifted_qe_d], color='blue')
        plt.text(initial_pe, initial_qe, f'Initial (Pe, Qe) = ({initial_pe:.2f}, {initial_qe:.2f})', ha='right')
        plt.text(shifted_pe_d, shifted_qe_d, f'Shifted (Pe, Qe) = ({shifted_pe_d:.2f}, {shifted_qe_d:.2f})', ha='right')
        plt.xlabel('Price')
        plt.ylabel('Quantity')
        plt.title('Equilibrium Shift')
        plt.grid(True)
        plt.show()
    
    if shifted_supply is not None:
        print("\nNew Equilibrium with Shifted Supply:")
        # Calculate and plot new equilibrium with shifted supply
        shifted_pe_s, shifted_qe_s = market_equilibrium(initial_demand, shifted_supply)
        
        # Plot equilibrium shift
        plt.figure(figsize=(10, 6))
        plt.plot([initial_pe, shifted_pe_s], [initial_qe, shifted_qe_s], marker='o')
        plt.scatter([initial_pe, shifted_pe_s], [initial_qe, shifted_qe_s], color='blue')
        plt.text(initial_pe, initial_qe, f'Initial (Pe, Qe) = ({initial_pe:.2f}, {initial_qe:.2f})', ha='right')
        plt.text(shifted_pe_s, shifted_qe_s, f'Shifted (Pe, Qe) = ({shifted_pe_s:.2f}, {shifted_qe_s:.2f})', ha='right')
        plt.xlabel('Price')
        plt.ylabel('Quantity')
        plt.title('Equilibrium Shift')
        plt.grid(True)
        plt.show()
```

```python
initial_demand = [(10, 80), (12, 70), (14, 60), (16, 50), (18, 40)]
initial_supply = [(10, 30), (12, 40), (14, 50), (16, 60), (18, 70)]
shifted_demand = [(10, 100), (12, 90), (14, 80), (16, 70), (18, 60)]  # Demand shift to the right
shifted_supply = [(10, 20), (12, 30), (14, 40), (16, 50), (18, 60)]   # Supply shift to the left
```

```python
plot_equilibrium_shift(initial_demand, initial_supply, shifted_demand, shifted_supply)
```
