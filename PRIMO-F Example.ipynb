```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## People 

```python
def cultural_diversity_index(employees_with_int_exp, total_employees):
    percentage = (employees_with_int_exp / total_employees) * 100
    if 0 <= percentage <= 25:
        return 0
    elif 26 <= percentage <= 50:
        return 1
    elif 51 <= percentage <= 75:
        return 2
    elif 76 <= percentage <= 100:
        return 3

def training_investment_per_employee(total_training_budget, total_employees):
    investment = total_training_budget / total_employees
    if 0 <= investment <= 500:
        return 0
    elif 501 <= investment <= 1000:
        return 1
    elif 1001 <= investment <= 1500:
        return 2
    elif investment >= 1501:
        return 3

def global_team_expertise_index(employees_with_expertise, total_employees):
    percentage = (employees_with_expertise / total_employees) * 100
    if 0 <= percentage <= 20:
        return 0
    elif 21 <= percentage <= 40:
        return 1
    elif 41 <= percentage <= 60:
        return 2
    elif 61 <= percentage <= 80:
        return 3
    elif 81 <= percentage <= 100:
        return 4
```

## Resources

```python
def technological_readiness(tech_score):
    return tech_score

def supply_chain_coverage(regions_covered, total_regions):
    percentage = (regions_covered / total_regions) * 100
    if 0 <= percentage <= 20:
        return 0
    elif 21 <= percentage <= 40:
        return 1
    elif 41 <= percentage <= 60:
        return 2
    elif 61 <= percentage <= 80:
        return 3
    elif 81 <= percentage <= 100:
        return 4

def local_market_data_access(access_level):
    return access_level
```

## Innovation

```python
def r_and_d_investment_percentage(r_and_d_expenditure, total_revenue):
    percentage = (r_and_d_expenditure / total_revenue) * 100
    if 0 <= percentage <= 2:
        return 0
    elif 3 <= percentage <= 5:
        return 1
    elif 6 <= percentage <= 8:
        return 2
    elif 9 <= percentage <= 12:
        return 3
    elif percentage >= 12:
        return 4

def new_product_launches(num_launches):
    if num_launches == 0:
        return 0
    elif num_launches == 1:
        return 1
    elif num_launches == 2:
        return 2
    elif num_launches >= 3:
        return 3

def speed_of_market_adaptation(months):
    if months == 12:
        return 0
    elif 9 <= months < 12:
        return 1
    elif 6 <= months < 9:
        return 2
    elif months < 6:
        return 3
```

## Marketing

```python
def marketing_roi(revenue, spend):
    roi = revenue / spend
    if roi < 1:
        return 0
    elif 1 <= roi <= 2:
        return 1
    elif 2 < roi <= 4:
        return 2
    elif roi > 4:
        return 3

def brand_recognition(awareness_percentage):
    if 0 <= awareness_percentage <= 20:
        return 0
    elif 21 <= awareness_percentage <= 40:
        return 1
    elif 41 <= awareness_percentage <= 60:
        return 2
    elif 61 <= awareness_percentage <= 80:
        return 3
    elif 81 <= awareness_percentage <= 100:
        return 4

def market_interest_rate(pre_launch_interactions, total_customers):
    percentage = (pre_launch_interactions / total_customers) * 100
    if 0 <= percentage <= 25:
        return 0
    elif 26 <= percentage <= 50:
        return 1
    elif 51 <= percentage <= 75:
        return 2
    elif 76 <= percentage <= 100:
        return 3
```

## Operations

```python
def cost_per_unit_production(international_cost, domestic_cost):
    if international_cost > domestic_cost:
        return 0
    elif international_cost == domestic_cost:
        return 1
    elif domestic_cost * 0.90 <= international_cost < domestic_cost:
        return 2
    elif domestic_cost * 0.80 <= international_cost < domestic_cost * 0.90:
        return 3
    return 3

def distribution_speed(average_delivery_time, competitor_delivery_time):
    if average_delivery_time < competitor_delivery_time:
        faster_percentage = ((competitor_delivery_time - average_delivery_time) / competitor_delivery_time) * 100
        if faster_percentage >= 20:
            return 4
        elif 10 <= faster_percentage < 20:
            return 3
    elif average_delivery_time == competitor_delivery_time:
        return 2
    return 0

def market_readiness_score(tasks_completed, total_tasks):
    if tasks_completed == 0:
        return 0
    elif 1 <= tasks_completed <= 2:
        return 1
    elif tasks_completed == 3:
        return 2
    elif tasks_completed == total_tasks:
        return 3
```

## Finance

```python
def international_revenue_growth_rate(current_revenue, last_revenue):
    if current_revenue < last_revenue:
        return 0
    growth = ((current_revenue - last_revenue) / last_revenue) * 100
    if 0 <= growth <= 10:
        return 1
    elif 11 <= growth <= 20:
        return 2
    elif growth >= 21:
        return 3

def currency_risk_management(hedged_percentage):
    if hedged_percentage == 0:
        return 0
    elif 0 < hedged_percentage < 25:
        return 1
    elif 25 <= hedged_percentage < 50:
        return 2
    elif 51 <= hedged_percentage < 75:
        return 3
    elif hedged_percentage >= 75:
        return 4

def budget_allocation_for_expansion(total_revenue, allocated_budget):
    percentage = (allocated_budget / total_revenue) * 100
    if 0 <= percentage <= 2:
        return 0
    elif 3 <= percentage <= 5:
        return 1
    elif 6 <= percentage <= 8:
        return 2
    elif percentage >= 9:
        return 3
```

##  Provided Data

```python
data = {
    'People': {
        'total_employees': 100,
        'employees_with_int_exp': 40,
        'total_training_budget': 80000,  # $80,000
        'employees_related_to_training': 100,
        'employees_with_expertise': 25
    },
    'Resources': {
        'tech_readiness_score': 3,  # Fully developed international systems
        'supply_chain_covered_regions': 5,
        'total_supply_chain_regions': 10,
        'local_market_data_access': 3  # Primary + Secondary data
    },
    'Innovation': {
        'total_revenue': 2000000,  # $2,000,000
        'r_and_d_expenditure': 180000,  # $180,000
        'new_product_launches': 2,
        'market_adaptation_time_months': 8
    },
    'Marketing': {
        'marketing_revenue': 300000,  # $300,000
        'marketing_spend': 100000,  # $100,000
        'brand_awareness_percentage': 60,
        'pre_launch_interactions': 1200,
        'total_customers': 3000
    },
    'Operations': {
        'international_cost_per_unit': 80,  # $80
        'domestic_cost_per_unit': 90,  # $90
        'average_delivery_time': 5,  # days
        'competitor_delivery_time': 6.25,  # assuming competitor takes 6.25 days for 20% faster
        'tasks_completed': 3,
        'total_tasks': 4
    },
    'Finance': {
        'current_international_revenue': 500000,  # $500,000
        'last_international_revenue': 400000,  # $400,000
        'currency_hedged_percentage': 30,
        'total_revenue': 2000000,  # $2,000,000
        'budget_allocation_expansion': 100000  # $100,000
    }
}
```

##  Calculate Scores

```python
# People
people_cdi = cultural_diversity_index(
    data['People']['employees_with_int_exp'],
    data['People']['total_employees']
)
people_tipe = training_investment_per_employee(
    data['People']['total_training_budget'],
    data['People']['employees_related_to_training']
)
people_gtei = global_team_expertise_index(
    data['People']['employees_with_expertise'],
    data['People']['total_employees']
)
people_score = people_cdi + people_tipe + people_gtei

# Resources
resources_tr = technological_readiness(
    data['Resources']['tech_readiness_score']
)
resources_scc = supply_chain_coverage(
    data['Resources']['supply_chain_covered_regions'],
    data['Resources']['total_supply_chain_regions']
)
resources_lmda = local_market_data_access(
    data['Resources']['local_market_data_access']
)
resources_score = resources_tr + resources_scc + resources_lmda

# Innovation
innovation_r_and_d = r_and_d_investment_percentage(
    data['Innovation']['r_and_d_expenditure'],
    data['Innovation']['total_revenue']
)
innovation_npl = new_product_launches(
    data['Innovation']['new_product_launches']
)
innovation_sma = speed_of_market_adaptation(
    data['Innovation']['market_adaptation_time_months']
)
innovation_score = innovation_r_and_d + innovation_npl + innovation_sma

# Marketing
marketing_roi_score = marketing_roi(
    data['Marketing']['marketing_revenue'],
    data['Marketing']['marketing_spend']
)
marketing_br = brand_recognition(
    data['Marketing']['brand_awareness_percentage']
)
marketing_mir = market_interest_rate(
    data['Marketing']['pre_launch_interactions'],
    data['Marketing']['total_customers']
)
marketing_score = marketing_roi_score + marketing_br + marketing_mir

# Operations
operations_cpp = cost_per_unit_production(
    data['Operations']['international_cost_per_unit'],
    data['Operations']['domestic_cost_per_unit']
)
operations_ds = distribution_speed(
    data['Operations']['average_delivery_time'],
    data['Operations']['competitor_delivery_time']
)
operations_mrs = market_readiness_score(
    data['Operations']['tasks_completed'],
    data['Operations']['total_tasks']
)
operations_score = operations_cpp + operations_ds + operations_mrs

# Finance
finance_irgr = international_revenue_growth_rate(
    data['Finance']['current_international_revenue'],
    data['Finance']['last_international_revenue']
)
finance_crm = currency_risk_management(
    data['Finance']['currency_hedged_percentage']
)
finance_bae = budget_allocation_for_expansion(
    data['Finance']['total_revenue'],
    data['Finance']['budget_allocation_expansion']
)
finance_score = finance_irgr + finance_crm + finance_bae

# Compile Scores
scores = {
    'People': people_score,
    'Resources': resources_score,
    'Innovation': innovation_score,
    'Marketing': marketing_score,
    'Operations': operations_score,
    'Finance': finance_score
}
```

## Visualization

```python
print("PRIMO-F Analysis Scores for TechWave Inc.\n")
print("1. People:")
print(f"   - Cultural Diversity Index: {people_cdi} points")
print(f"   - Training Investment per Employee: {people_tipe} points")
print(f"   - Global Team Expertise Index: {people_gtei} points")
print(f"   **Total People Score:** {people_score}\n")

print("2. Resources:")
print(f"   - Technological Readiness: {resources_tr} points")
print(f"   - Supply Chain Coverage: {resources_scc} points")
print(f"   - Local Market Data Access: {resources_lmda} points")
print(f"   **Total Resources Score:** {resources_score}\n")

print("3. Innovation:")
print(f"   - R&D Investment as % of Revenue: {innovation_r_and_d} points")
print(f"   - New Product Launches: {innovation_npl} points")
print(f"   - Speed of Market Adaptation: {innovation_sma} points")
print(f"   **Total Innovation Score:** {innovation_score}\n")

print("4. Marketing:")
print(f"   - Marketing ROI: {marketing_roi_score} points")
print(f"   - Brand Recognition in International Market: {marketing_br} points")
print(f"   - Market Interest Rate: {marketing_mir} points")
print(f"   **Total Marketing Score:** {marketing_score}\n")

print("5. Operations:")
print(f"   - Cost per Unit of Production: {operations_cpp} points")
print(f"   - Distribution Speed: {operations_ds} points")
print(f"   - Market Readiness Score: {operations_mrs} points")
print(f"   **Total Operations Score:** {operations_score}\n")

print("6. Finance:")
print(f"   - International Revenue Growth Rate: {finance_irgr} points")
print(f"   - Currency Risk Management: {finance_crm} points")
print(f"   - Budget Allocation for Expansion: {finance_bae} points")
print(f"   **Total Finance Score:** {finance_score}\n")

##### Radar Chart Visualization

categories = list(scores.keys())
N = len(categories)

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Complete the loop

values = list(scores.values())
values += values[:1]

plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1], categories, color='grey', size=12)

ax.set_rlabel_position(30)
plt.yticks([2,4,6,8,10], ["2","4","6","8","10"], color="grey", size=10)
plt.ylim(0,10)

ax.plot(angles, values, linewidth=1, linestyle='solid', label='TechWave Inc.')

ax.fill(angles, values, 'skyblue', alpha=0.4)

for angle, score, category in zip(angles, values, categories):
    angle_rad = angle
    if angle_rad >= np.pi / 2 and angle_rad <= 3 * np.pi / 2:
        alignment = "right"
    else:
        alignment = "left"
    ax.text(angle_rad, score + 0.5, f"{score}", size=12, horizontalalignment=alignment, verticalalignment='center')

plt.title('PRIMO-F Internal Analysis of TechWave Inc.', size=15, y=1.08)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
