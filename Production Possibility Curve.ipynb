```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Production cap
production_cap = 14

# Production scenarios (A to E)
scenarios = {
    'A': (2, 12),
    'B': (4, 10),
    'C': (6, 8),
    'D': (8, 6),
    'E': (10, 4)
}
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Vizualisation
</h1><center/>

```python
# Extracting data for plotting
product_a = [scenarios[scenario][0] for scenario in scenarios]
product_b = [scenarios[scenario][1] for scenario in scenarios]

# PPC Line
x = np.linspace(0, production_cap, 100)
y = production_cap - x

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='PPC Line (Optimal Production)', color='green')
plt.scatter(product_a, product_b, color='red')

# Annotating points
for scenario in scenarios:
    plt.annotate(scenario, (scenarios[scenario][0], scenarios[scenario][1]), textcoords="offset points", xytext=(0,10), ha='center')

# Adding labels and title
plt.xlabel('Product A')
plt.ylabel('Product B')
plt.title('Production Possibility Curve (PPC)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

<center><h1 style="background-color: #C6F3CD; border-radius: 10px; color: #FFFFFF; padding: 5px;">
Function
</h1><center/>

```python
def plot_ppc_and_calculate_tradeoffs(production_cap, scenarios):
    # Extracting data for plotting
    product_a = [scenarios[scenario][0] for scenario in scenarios]
    product_b = [scenarios[scenario][1] for scenario in scenarios]

    # PPC Line
    x = np.linspace(0, production_cap, 100)
    y = production_cap - x

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='PPC Line (Optimal Production)', color='green')
    plt.scatter(product_a, product_b, color='red')

    # Annotating points
    for scenario in scenarios:
        plt.annotate(scenario, (scenarios[scenario][0], scenarios[scenario][1]), textcoords="offset points", xytext=(0,10), ha='center')

    # Adding labels and title
    plt.xlabel('Product A')
    plt.ylabel('Product B')
    plt.title('Production Possibility Curve (PPC)')
    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.show()

    # Creating DataFrame for optimal points
    data = []
    for scenario in scenarios:
        a = scenarios[scenario][0]
        b = scenarios[scenario][1]
        is_optimal = 'Yes' if a + b == production_cap else 'No'
        data.append({
            'Scenario': scenario,
            'Product A': a,
            'Product B': b,
            'Optimal Production': is_optimal
        })

    df = pd.DataFrame(data)
    print(df)
```

```python
# Production cap
production_cap = 14

# Production scenarios (A to E)
scenarios = {
    'A': (2, 12),
    'B': (4, 8),
    'C': (6, 8),
    'D': (8, 6),
    'E': (10, 4)
}

# Calling the function
plot_ppc_and_calculate_tradeoffs(production_cap, scenarios)
```
