```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data: brand names, prices, and healthiness scores
brands = ['Brand A', 'Brand B', 'Brand C', 'Brand D', 'Brand E']
prices = [3, 5, 8, 2, 4]  # Price scale (low to high)
healthiness = [6, 7, 4, 5, 8]  # Healthiness scale (low to high)

# Convert prices and healthiness into numpy arrays for easier manipulation
prices = np.array(prices)
healthiness = np.array(healthiness)

# Step 3: Create the Positioning Map
plt.figure(figsize=(10, 8))
plt.scatter(prices, healthiness, s=100, color='blue')

# Annotate the brands
for i, brand in enumerate(brands):
    plt.annotate(brand, (prices[i], healthiness[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Add midlines for better interpretation
plt.axhline(y=5, color='r', linestyle='--', label='Average Healthiness')  
plt.axvline(x=5, color='g', linestyle='--', label='Average Price')  

# Title and labels
plt.title('Positioning Map of Health Food Brands')
plt.xlabel('Price (Low to High)')
plt.ylabel('Healthiness (Low to High)')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.legend()
plt.show()
```
